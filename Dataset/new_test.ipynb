```python

def calculate_score(row):
    if not row['rule_3'] or not row['rule_4']:
        return 0
    return sum([row[f'rule_{i}'] for i in range(1, 10)])

df['ai_score'] = df.apply(calculate_score, axis=1)
```

```python
import pandas as pd

# Load the CSV files
df_ai = pd.read_csv("ai_grade_copy.csv")
df_human = pd.read_csv("human_reasoning.csv")

# Normalize column names
df_ai.columns = df_ai.columns.str.strip().str.lower()
df_human.columns = df_human.columns.str.strip().str.lower()

# Merge on image_id and id
merged_df = pd.merge(df_ai, df_human, left_on="image_id", right_on="image_id", how="inner")

# Filter out rows where rule_3 or rule_4 is True (skip rule failures)
filtered_df = merged_df[
    (merged_df.get('rule_3').fillna(0) == 0) & 
    (merged_df.get('rule_4').fillna(0) == 0)
]

# Compute AI score by summing rule_1 to rule_9
rule_cols = [f'rule_{i}' for i in range(1, 10) if f'rule_{i}' in filtered_df.columns]
filtered_df['ai_score'] = filtered_df[rule_cols].sum(axis=1)

# Compare AI score to human grade
filtered_df['score_diff'] = filtered_df['human_grade'] - filtered_df['ai_score']

# Output stats
mean_diff = filtered_df['score_diff'].mean()
std_diff = filtered_df['score_diff'].std()

print("Mean difference:", mean_diff)
print("Standard deviation:", std_diff)
print(filtered_df)
```
